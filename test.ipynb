```python
### import libraries
import os
import tensorflow as tf
import numpy as np
```

```python
## DATASET
datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)



test_path = 'test/'
test_generator = datagen.flow_from_directory(test_path, shuffle=True, target_size=(224,224))
```

```python
model = tf.keras.models.load_model('./models/model_1/')
model.summary()
```

```python
pred=model.predict_generator(test_generator,verbose=1)
```

```python
predicted_class_incdices = np.argmax(pred, axis=1)
```

```python
labels = (test_generator.class_indices)
labels = dict((v,k) for k,v in labels.items())
labels
```

```python
predictions = [labels[p] for p in predicted_class_incdices]
```
